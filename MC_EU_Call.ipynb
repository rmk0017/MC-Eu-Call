# Monte Carlo Simulation to value European Style Call Options 

```python
#
# # Import Relevant Libraries
#
import math

import numpy as np
import matplotlib.pyplot as plt
```

```python
#
# # Parameters
#   n: number of simulations
#   t: time to maturity
#   r: risk free rate
#   v: volatility
#   S_0: initial stock price
#   K: strike price
#
n = 100000
t = 1
r = 0.05
v = 0.25
S_0 = 105
K = 100

#
# # Generate a list of n random numbers
#
w = np.random.standard_normal(n)

#
# # Compute possible future values of the stock based on geometric brownian motion
#
S_T = S_0 * (np.exp( (r - (v ** 2) / 2) * t + v * w * np.sqrt(t)))

#
# # Computing the price of the call option (taking the average price of the call option)
#
call = (np.sum(np.maximum(S_T - K, 0) * (np.exp(-r * t)))) / n
call
```

```python
#
# # Plot values of the stock in a histogram
#
plt.hist(S_T, 50)
plt.xlabel("Stock Price")
plt.ylabel("Frequency")
plt.title('Monte Carlo Simulation of Stock Prices')
plt.axvline(S_T.mean(), color='r', linestyle='dashed', linewidth=1)
plt.show()
```
